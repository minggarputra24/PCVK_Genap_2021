```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/PCVK/Images/coin1.jpg', 0)
blur = cv2.GaussianBlur(img, (11, 11), 0)
edges = cv2.Canny(blur, 40, 200)
dilated = cv2.dilate(edges, (1,1), iterations = 2)

(cnt, _) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print("Terdapat {} Koin dalam gambar".format(len(cnt)))

plt.imshow(dilated)

# plt.waitKey(0)
# plt.destroyAllWindows()
```
